## 2.1 Defining scraping and extraction functions

The list of URLs to be analysed during this project is in a CSV file. Prior to importing this file and scraping the content of the URLs, we need to define a function which we can call to scrape the data as we iterate through the URLs. Each website we are going to scrape from has different HTML, so we also need to write a different function to extract the required information from each.

The below function fetches a webpage and returns a BeautifulSoup object for HTML parsing.

```python
def getParsedWebpage(url, website):
    # Headers to mimic a real browser visit
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36",
        "Upgrade-Insecure-Requests": "1",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8",
        "Accept-Encoding": "gzip, deflate, br",
        "Accept-Language": "en-GB,en-US;q=0.9,en;q=0.8",
        "DNT": "1"
    }

    # Optional delay for specific websites
    if website == 'mirror':
        time.sleep(12)

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Raise HTTPError for bad responses
        content = response.content

        # Log content preview for debugging (optional)
        print(content[:500])  # Print the first 500 characters to check the content
        
        # Parse the webpage content
        soup = BeautifulSoup(content, 'html.parser')
    except requests.exceptions.RequestException as e:
        print(f"Error fetching the page: {e}")
        soup = 'error'

    return soup
```

### 2.1.2 Use a BeautifulSoup method for extracting data


I am using find() and find_all() methods of the BeautifulSoup object to locate the table body and table row respectively in the HTML.

The find() method will return particular tag content.
The find_all() method returns a list of all matching tags in the HTML.

```python
# Function to parse table data
def parseTableData(soup):
    # Initialize an empty list to store row data
    rows_data = []

    # Find the table body
    tbody = soup.find("tbody")
    if not tbody:
        print("No table body found")
        return pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"])
    
    # Iterate through each row of the table
    for row in tbody.find_all('tr'):
        col = row.find_all("td")
        
        # Ensure the row has enough columns
        if len(col) != 7:
            print(f"Skipping row due to incorrect number of columns: {len(col)}")
            continue  # Skip this row if it doesn't have exactly 7 columns
        
        # Extract and clean text from each column
        date = col[0].text.strip()
        open_price = col[1].text.strip()
        high = col[2].text.strip()
        low = col[3].text.strip()
        close = col[4].text.strip()
        adj_close = col[5].text.strip()
        volume = col[6].text.strip().replace(',', '')  # Remove commas from volume

        # Append the row data as a dictionary to the list
        rows_data.append({
            "Date": date,
            "Open": open_price,
            "High": high,
            "Low": low,
            "Close": close,
            "Adj Close": adj_close,
            "Volume": volume
        })

    # Convert the list of dictionaries to a DataFrame
    tesla_data = pd.DataFrame(rows_data)
    
    return tesla_data


# Display the DataFrame
print(tesla_data)
```

### 2.2.2 Scrape data and save to dataframe
Next we will attempt to scrape the articles at each of the URLs we have imported using the functions defined earlier. If successful, we will append to a dataframe of scraped information.

```python
# Function to fetch and parse webpage content
def getParsedWebpage(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36",
        "Upgrade-Insecure-Requests": "1",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8",
        "Accept-Encoding": "gzip, deflate, br",
        "Accept-Language": "en-GB,en-US;q=0.9,en;q=0.8",
        "DNT": "1"
    }

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')
    except requests.exceptions.RequestException as e:
        print(f"Error fetching the page: {e}")
        return 'error'

    return soup

# Load URLs from CSV into a DataFrame
urls = pd.read_csv("urls.csv")

# Initialize an empty DataFrame to store scraped data
scraped_data = pd.DataFrame(columns=["Name", "URL"])

# Loop through each URL in the DataFrame
for i in range(len(urls)):
    url = urls.loc[i, 'URL']
    publication = urls.loc[i, 'Name']
    
    # Fetch and parse the webpage
    soup = getParsedWebpage(url)
    
    # If there was an error accessing the webpage, we can skip to the next iteration
    if soup == 'error':
        print("Error fetching webpage, data not scraped:", url)
        continue
    
    # Append the scraped data to the DataFrame
    scraped_data = scraped_data.append({'Name': publication, 'URL': url}, ignore_index=True)

# Display the scraped data DataFrame
print(scraped_data)

# Save the scraped data to a CSV file
scraped_data.to_csv("scraped_data.csv", index=False)
```

### 3.2.1 Identifying stop words not covered by NLTK's stop words
Stop words can be removed using nltk's stop words. However, there may be other words in the corpus which can be functionally considered stop words and which are of little use to us in comparative analysis. To spot some such words and remove them, we first analyse frequency of terms in each corpus, to see which are most common:

```python
# First I define a function to get the word frquencies of each corpus
# At this stage I will also remove stop words as defined by nltk's stopwords set

nltk.download('stopwords')
```

```python
# Example DataFrame (replace with your actual compiled data)
compiled_corpus = pd.DataFrame({
    'Publication': ['barchart', 'Trading Economics'],
    'Full corpus': [
        "Full text of all articles from Barchart",
        "Full text of all articles from Trading Economics"
    ]
})

# Function to calculate word frequencies
def getFreqs(corpus):
    # Remove certain punctuation using regex (preserve apostrophes)
    corpus = re.sub(r'\.|\,|\"|\“|\”|\—|\–|\-|\?|\!|\:|\;|\(|\)', '', corpus)
    
    # Split the corpus into a list of words
    corpus_lst = corpus.lower().split()
    
    # Remove stop words
    filtered_words = [word for word in corpus_lst if word not in stopwords.words('english')]
    
    # Calculate frequencies
    word_freqs = pd.Series(filtered_words).value_counts().reset_index()
    word_freqs.columns = ['Word', 'Frequency']
    
    return word_freqs

# Generate word frequencies for each publication
barchart_word_freqs = getFreqs(compiled_corpus['Full corpus'][0])
trading_economics_word_freqs = getFreqs(compiled_corpus['Full corpus'][1])

# Display word frequencies
print("Barchart Word Frequencies:\n", barchart_word_freqs)
print("\nTrading Economics Word Frequencies:\n", trading_economics_word_freqs)
```

## 3.3 Lemmatization
We can 'distill' the words we have into their roots to make them more meaningful for analysis. We can either use stemming or lemmatization. Stemming removes suffixes to create a root stem (so 'studies' would become 'stud'), while lemmatization will return a root word ('study').

In our later analysis, we want to be able to see whole, meaningful words and make impactful visualisations using them. Therefore in this case it is better to use lemmatization to get root words.

Because creating the root of words may sometimes give us unexpected results, we will create a separate column for lemmatized corpora. We'll want to use the filtered corpora with stop words removed, but without lemmatization, for some of our exploratory data analysis later.

Below we create an instance of the wordnet lemmatizer, create a lemmatizing function and apply to the filtered corpora:

```python
%pip install wordcloud
```

```python
# Download the required NLTK resources
nltk.download('wordnet')
nltk.download('omw-1.4')

# Initialize the lemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

# Example DataFrame
df = pd.DataFrame({
    'Text': [
        "The cats are running in the gardens.",
        "He enjoys reading books and walking."
    ]
})

# Define the lemmatization function
def lemmatize_text(text):
    # Remove punctuation, split text into words
    words = re.sub(r'[^\w\s]', '', text).split()
    # Lemmatize each word
    lemmatized = [wordnet_lemmatizer.lemmatize(word) for word in words]
    # Join words back into a string
    return ' '.join(lemmatized)

# Apply lemmatization to the DataFrame
df['Lemmatized Text'] = df['Text'].apply(lemmatize_text)

# Display the DataFrame
print(df)
```

```python
# Load CSV file into a DataFrame (replace 'TSLA.csv' with your actual file path)
df = pd.read_csv('TSLA.csv')

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Calculate daily returns
df['Daily Return'] = df['Close'].pct_change()

# Calculate moving averages
df['20-Day MA'] = df['Close'].rolling(window=20).mean()
df['50-Day MA'] = df['Close'].rolling(window=50).mean()

# Calculate Bollinger Bands
df['20-Day STD'] = df['Close'].rolling(window=20).std()
df['Upper Band'] = df['20-Day MA'] + (df['20-Day STD'] * 2)
df['Lower Band'] = df['20-Day MA'] - (df['20-Day STD'] * 2)

# Define a function to calculate RSI
def calculate_rsi(data, window):
    delta = data['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

# Calculate 14-day RSI
df['14-Day RSI'] = calculate_rsi(df, window=14)

# Plot Close price and Moving Averages
plt.figure(figsize=(14,7))
plt.plot(df['Date'], df['Close'], label='Close Price')
plt.plot(df['Date'], df['20-Day MA'], label='20-Day MA', linestyle='--')
plt.plot(df['Date'], df['50-Day MA'], label='50-Day MA', linestyle='--')
plt.title('Stock Price and Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Plot Bollinger Bands
plt.figure(figsize=(14,7))
plt.plot(df['Date'], df['Close'], label='Close Price')
plt.plot(df['Date'], df['Upper Band'], label='Upper Band', linestyle='--')
plt.plot(df['Date'], df['Lower Band'], label='Lower Band', linestyle='--')
plt.title('Bollinger Bands')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Plot RSI
plt.figure(figsize=(14,7))
plt.plot(df['Date'], df['14-Day RSI'], label='14-Day RSI', linestyle='--')
plt.title('Relative Strength Index')
plt.xlabel('Date')
plt.ylabel('RSI')
plt.legend()
plt.show()
```

### 4.1.1 Interpret Sentiment Scores
The VADER analyzer provides the following scores:

compound: Overall sentiment score (normalized between -1 and 1)
neg: Negative sentiment proportion
neu: Neutral sentiment proportion
pos: Positive sentiment proportion

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download VADER lexicon
nltk.download('vader_lexicon')

# Initialize VADER sentiment intensity analyzer
sia = SentimentIntensityAnalyzer()

# Load data from CSV file
df = pd.read_csv('TSLA.csv')

# Example data
data = {
    'Date': ['2024-06-01', '2024-06-02'],
    'Text': ['The stock market is soaring. Great time to invest!', 'The stock market is crashing. Terrible losses ahead.']
}
df = pd.DataFrame(data)

# Define function to analyze sentiment
def analyze_sentiment(text):
    return sia.polarity_scores(text)

# Apply sentiment analysis to each row in the 'Text' column
df['Sentiment'] = df['Text'].apply(analyze_sentiment)

# Separate the sentiment scores into individual columns
df = df.join(df['Sentiment'].apply(pd.Series))

# Display the DataFrame with sentiment scores
print(df[['Date', 'Text', 'compound', 'neg', 'neu', 'pos']])
```

### 5.2.2 Correlation and Causation
Correlation: a measure of the extent of interdependence between variables.

Causation: the relationship between cause and effect between two variables.

It is important to know the difference between these two. Correlation does not imply causation. Determining correlation is much simpler the determining causation as causation may require independent experimentation.

Pearson Correlation

The Pearson Correlation measures the linear dependence between two variables X and Y.

The resulting coefficient is a value between -1 and 1 inclusive, where:

1: Perfect positive linear correlation.
0: No linear correlation, the two variables most likely do not affect each other.
-1: Perfect negative linear correlation.
